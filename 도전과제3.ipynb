```python
import pandas as pd  # 판다스 라이브러리 가져오기

file = "/Users/josuhyeon/Downloads/pop_kor.csv"  # 파일 경로 설정
# CSV 파일 불러오기
df_pop = pd.read_csv(file, index_col='구별')  # '구별'을 DataFrame의 인덱스로 설정
print(df_pop)  # 데이터프레임 출력
```

```python
crime_file = "/Users/josuhyeon/Downloads/관서별 5대범죄 발생 및 검거.xlsx"  # 1. 범죄 데이터 불러오기
df_crime = pd.read_excel(crime_file)  # 엑셀 파일 읽기

police_to_gu = {  # '관서명' -> '구별' 매핑
    '서대문서': '서대문구', '수서서': '강남구', '강서서': '강서구', '서초서': '서초구',
    '서부서': '은평구', '중부서': '중구', '종로서': '종로구', '남대문서': '중구',
    '혜화서': '종로구', '용산서': '용산구', '성북서': '성북구', '동대문서': '동대문구',
    '마포서': '마포구', '영등포서': '영등포구', '성동서': '성동구', '동작서': '동작구',
    '광진서': '광진구', '강북서': '강북구', '금천서': '금천구', '중랑서': '중랑구',
    '강남서': '강남구', '관악서': '관악구', '강동서': '강동구', '종암서': '성북구', 
    '구로서': '구로구', '양천서': '양천구', '송파서': '송파구', '노원서': '노원구', 
    '방배서': '서초구', '은평서': '은평구', '도봉서': '도봉구'
}
df_crime['구별'] = df_crime['관서명'].map(police_to_gu).fillna('구 없음')  # 구별 컬럼 생성
df_crime.set_index('구별', inplace=True)  # '구별'을 인덱스로 설정
print("Crime DataFrame:")
print(df_crime)

# 2. 인구 데이터 불러오기
pop_file = "/Users/josuhyeon/Downloads/pop_kor.csv"
df_pop = pd.read_csv(pop_file, index_col='구별')  # '구별'을 인덱스로 설정
print("\nPopulation DataFrame:")
print(df_pop)

# 3. 병합
df_merged = df_crime.join(df_pop, how='inner')  # inner join: 공통된 '구별' 인덱스만 병합
print("모두합쳐")
print(df_merged)
```

```python
# 검거율 계산
# '검거율' 컬럼 추가 (소계(검거) / 소계(발생) * 100)
df_merged['검거율'] = (df_merged['소계(검거)'] / df_merged['소계(발생)']) * 100

# 오름차순 정렬
df_sorted = df_merged.sort_values(by='검거율', ascending=True)  # '검거율' 기준으로 오름차순 정렬

# 결과 출력
print("검거율 오름차순 정렬")
print(df_sorted)
```
